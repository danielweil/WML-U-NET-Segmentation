```python
import SimpleITK as sitk
import os
from pylab import *
from pathlib import Path
import nibabel as nib
import pydicom
from fastai.basics import *
from fastai.callback.all import *
from fastai.vision.all import *
from fastai.medical.imaging import *
```

```python
def showImages(source):
    plt.imshow(source, cmap="gray")
    plt.show()
```

```python
def plotHistograms(source):
    plt.hist(source.ravel(), bins=np.arange(1, 4000, 1), fc='k', ec='b');
    plt.tight_layout()
    plt.show()
```

```python
def iterateNiftiExam(examNumber):
    source = nib.load(f'C:/Users/Visagio/Desktop/dt/skullstripped/{examNumber}/masked.nii.gz')
    exam = source.get_fdata()
    for sliceN in range(0, exam.shape[2] - 1):
        showImages(np.rot90(exam[:,:,sliceN]))
        plotHistograms(np.rot90(exam[:,:,sliceN]))
```
