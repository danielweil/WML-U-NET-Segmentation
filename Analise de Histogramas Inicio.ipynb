```python
def showImages(source):
    plt.imshow(source, cmap="gray")
    plt.show()
```

```python
def plotHistograms(source):
    plt.hist(source.ravel(), bins=np.arange(1, 4000, 1), fc='k', ec='b');
    plt.tight_layout()
    plt.show()
```

```python
def iterateExamSlices(examNumber):
    items = get_dicom_files(r"C:\Users\Visagio\Desktop\dt\datasets", recurse=True, folders=str(examNumber))
    for item in items:
        exam = pydicom.dcmread(item).pixel_array
        showImages(exam)
        plotHistograms(exam)
```
